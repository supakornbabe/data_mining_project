```python
import json
import os
```

```python
# Load Data to data_store
```

```python
file_list=os.listdir("../Data")
print(file_list)

data_store=[]
for file in file_list:
    if file == ".DS_Store":
        continue
    print(file,end=" ")
    fo = open("../Data/" + file, "r")
    for line in fo:
        line = line.replace("{'", '{"')
        line = line.replace("'}", '"}')
        line = line.replace("':", '":')
        line = line.replace(":'", ':"')
        line = line.replace("',", '",')
        line = line.replace(",'", ',"')
        line = line.replace("['", '["')
        line = line.replace("']", '"]')
        line = line.replace(" '", ' "')
        
        print(line)
        data_store.append(json.loads(line))
    print("Finish")
```
